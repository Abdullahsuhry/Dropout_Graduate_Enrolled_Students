```python
import pandas as pd
```

```python
df = pd.read_csv('new11.csv')
```

```python
import matplotlib.pyplot as plt

target_col = "target"  # change if needed

df[target_col].value_counts().plot(kind='bar')
plt.title("Student Outcome Distribution")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns

df[num_cols].hist(bins=20, figsize=(15, 20))
plt.suptitle("Numerical Feature Distributions")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x=target_col, y="Age at enrollment", data=df)
plt.title("Boxplot of Age at Enrollment by Outcome")
plt.show()
```

```python
sns.countplot(x="Gender", hue=target_col, data=df)
plt.title("Gender vs Student Outcome")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[num_cols].corr(), cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.violinplot(x=target_col, y="Curricular units 1st sem (approved)", data=df)
plt.title("Approved Units vs Outcome")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = df.drop(columns=[target_col])
y = df[target_col]

X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=y)
plt.title("PCA Projection of Student Outcomes")
plt.show()
```
